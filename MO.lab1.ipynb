```python
from sklearn import ensemble
rf_model = ensemble.RandomForestClassifier(n_estimators = 10)
rf_model.fit(X_train, y_train)
```

```python
err_train = np.mean(y_train != rf_model.predict(X_train))
err_test  = np.mean(y_test  != rf_model.predict(X_test))
err_train, err_test
```

Заметим , что ошибка на тестовой выборке больше, чем на предыдущей. Значит наблюдаем переобучение.

```python

tree_arr = np.arange(1, 50, 10)
test_err = []
train_err = []
oob_err = []

for tree in tree_arr:
    rf_model = ensemble.RandomForestClassifier(n_estimators = tree, oob_score = True)
    rf_model.fit(X_train, y_train)
    y_train_pred = rf_model.predict(X_train)
    y_test_pred = rf_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - rf_model.oob_score_)
```

```python
plt.plot(tree_arr, train_err, 'b-o', label = 'train')
plt.plot(tree_arr, test_err, 'r-o', label = 'test')
plt.plot(tree_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(tree_arr), np.max(tree_arr)])
plt.title('Error vs. Count of tree')
plt.xlabel('Count of tree')
plt.ylabel('Error')
plt.legend()
pass
```

Заметим, что на всем диапазоне значений наблюдается ситуация переобучения, также видим, что oob ошибка превосходит ошибку на тестовой и обучающей выборках,за исключением  10-15 деревьев, где ошибка oob становится меньше значения ошибки на тестовой выборки. 

```python
tree_arr = np.arange(50, 150, 10)
test_err = []
train_err = []
oob_err = []

for tree in tree_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = tree, oob_score = True)
    rf.fit(X_train, y_train)
    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - rf.oob_score_)
```

```python
plt.plot(tree_arr, train_err, 'b-o', label = 'train')
plt.plot(tree_arr, test_err, 'r-o', label = 'test')
plt.plot(tree_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.min(tree_arr), np.max(tree_arr)])
plt.title('Error vs. Count of tree')
plt.xlabel('Count of tree')
plt.ylabel('Error')
plt.legend()
pass
```

Заметим, что процесс переобучения будет наблюдаться и далее, но ошибка на тестовой выборке будет уменьшаться с увеличением количества деревьев.

Минимальные значения на обучающей и тестовой выборках:

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

min_train_err,  min_test_err
```

Эти значения достигаются при следующем колчестве деревьев:

```python
trees_opt = tree_arr[test_err == min_test_err]
print(trees_opt)
```
